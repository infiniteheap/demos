# **Signal in Computer Networks**

### **What is a Signal?**

In computer networks, a signal represents how information is transmitted from one system to another. It is usually represented as a change in voltage over time. The signal starts at zero, increases until it peaks, and then drops back to its lowest level.

Signals can be in either analog or digital form and may take the form of electrical pulses, light, or electromagnetic waves. The pattern made by the signal is called a waveform, which can be a sine wave or a square wave.

### **Analog Signal**

An analog signal is a continuous waveform that changes smoothly over time. It can have an infinite number of values and is usually represented by a sine wave. An example of an analog signal is a human voice. When we speak, we use air to transmit an analog signal.

#### **Characteristics of Analog Signals:**

1. **Amplitude:**
   - The height of the signal, measured vertically from a point on the waveform to the horizontal axis.
   - Maximum amplitude corresponds to the highest value reached on the vertical axis.

2. **Period:**
   - The amount of time taken to complete one cycle of the signal.
   - Measured in seconds or milliseconds.

3. **Frequency:**
   - The number of cycles completed in one second.
   - Measured in Hertz (Hz).

4. **Phase:**
   - The position of the waveform relative to time zero.
   - Describes how much the waveform shifts forward or backward.

---

### **Advantages of Analog Signals**

1. Best suited for transmitting audio and video.
2. Consumes less bandwidth than digital signals.
3. Analog systems are widespread and readily available.
4. Less susceptible to noise.

---

### **Digital Signal**

A digital signal is a discrete waveform that can have only a limited number of defined values, typically 1 and 0. The transition between these values is instantaneous. Digital signals are represented by square waves. In digital signals:
- **1** is represented by a positive voltage.
- **0** is represented by zero voltage.

#### **Characteristics of Digital Signals:**

1. **Bit Interval:**
   - The time required to send one bit.

2. **Bit Rate:**
   - The number of bit intervals in one second.
   - Expressed in bits per second (bps).

---

### **Advantages of Digital Signals**

1. Best suited for transmitting digital data.
2. Digital data can be easily compressed and encrypted.
3. Equipment using digital signals is more common and affordable.
4. Provides better clarity since all signals must be either 1s or 0s.

---

### **Transmission of Digital Signals**

Digital signals can be transmitted using two main approaches:

1. **Baseband Transmission:**
   - Sending a digital signal without converting it to analog form.
   - Requires a low-pass channel.

2. **Broadband Transmission:**
   - Converts the digital signal to analog form using modulation.
   - Uses a bandpass channel for transmission.

---

### **Animated Demonstration of Analog and Digital Signals**

Here is an animated demonstration that highlights the key properties of both analog and digital signals.

#### **Analog Signal Properties:**
- **Amplitude:** The height of the wave.
- **Frequency:** The number of complete cycles per second.
- **Phase:** The shift along the time axis.

#### **Digital Signal Properties:**
- **Bit Rate:** The number of bits transmitted per second.
- **Bit Interval:** The time between each bit.

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time

# Use inline mode for Jupyter
%matplotlib inline

# Generate data for analog and digital signals
t = np.linspace(0, 2 * np.pi, 1000)
t_sampled = np.linspace(0, 2 * np.pi, 20)

# Create the figure and axis
fig, ax = plt.subplots(2, 1, figsize=(12, 10))

# Continuously update the plot to simulate animation
for i in range(100):
    # Clear previous plots
    ax[0].cla()
    ax[1].cla()

    # Set up the titles and limits again (since clearing resets them)
    ax[0].set_title('Analog Signal')
    ax[1].set_title('Digital Signal')

    ax[0].set_xlim(0, 2 * np.pi)
    ax[0].set_ylim(-1.5, 1.5)
    ax[1].set_xlim(0, 2 * np.pi)
    ax[1].set_ylim(-1.5, 1.5)

    # Update the analog signal (continuous)
    analog_signal = np.sin(2 * np.pi * (t + i * 0.1))  # Including phase shift
    ax[0].plot(t, analog_signal, color='blue')
    ax[0].text(0.02, 1.3, 'Properties:', transform=ax[0].transAxes, fontsize=12, fontweight='bold')
    ax[0].text(0.02, 1.2, 'Amplitude (Peak Value)', transform=ax[0].transAxes)
    ax[0].text(0.02, 1.1, 'Frequency (Oscillations per Second)', transform=ax[0].transAxes)
    ax[0].text(0.02, 1.0, 'Phase (Horizontal Shift)', transform=ax[0].transAxes)

    # Highlight the amplitude
    ax[0].plot([np.pi/2, np.pi/2], [0, 1], 'r--', label='Amplitude')
    ax[0].annotate('Amplitude', xy=(np.pi/2, 1), xytext=(np.pi/2 + 0.5, 1.2),
                   arrowprops=dict(facecolor='black', shrink=0.05))

    # Highlight the frequency (period)
    ax[0].plot([0, 2*np.pi], [0, 0], 'g--')
    ax[0].annotate('Frequency', xy=(np.pi, 0.2), xytext=(np.pi + 0.5, 0.5),
                   arrowprops=dict(facecolor='black', shrink=0.05))

    # Highlight the phase shift
    ax[0].annotate('Phase Shift', xy=(0, np.sin(i * 0.1)), xytext=(0.5, np.sin(i * 0.1) + 0.5),
                   arrowprops=dict(facecolor='black', shrink=0.05))

    # Update the digital signal (discrete)
    digital_signal = np.round(np.sin(2 * np.pi * (t_sampled + i * 0.1)))  # Including phase shift
    ax[1].plot(t_sampled, digital_signal, 'ro-')
    ax[1].text(0.02, 1.3, 'Properties:', transform=ax[1].transAxes, fontsize=12, fontweight='bold')
    ax[1].text(0.02, 1.2, 'Bit Rate (Bits per Second)', transform=ax[1].transAxes)
    ax[1].text(0.02, 1.1, 'Bit Interval (Time between Bits)', transform=ax[1].transAxes)

    # Highlight the bit interval
    for j in range(len(t_sampled) - 1):
        ax[1].plot([t_sampled[j], t_sampled[j]], [-1.5, 1.5], 'g--')
    ax[1].annotate('Bit Interval', xy=(t_sampled[1], 1), xytext=(t_sampled[1] + 0.5, 1.2),
                   arrowprops=dict(facecolor='black', shrink=0.05))

    # Display the plot and pause to simulate animation
    display(fig)
    clear_output(wait=True)
    time.sleep(0.05)  # Adjust the speed of updates for smoother animation

# Keep the last frame displayed
display(fig)
```
